```python
#ALGORITMO DE PESQUISA EM PROFUNDIDADE

def dfs(estado_inicial, bloco_final):
    start = time.time()
    pilha= deque()
    pilha.appendleft(([estado_inicial],[]))
    visitado = set()
    while pilha:
        atual,direcoes = pilha.pop()
        estado = atual[-1]
        if estado == bloco_final:
            end = time.time()
            print(f"Tempo: {round((end - start)*1000, 3)} ms")
            return atual,direcoes
        if str(estado) in visitado:
            continue
        visitado.add(str(estado))
        lista_filhos,lista_direcoes = filhos(estado)
        filho_direcao = list(zip(lista_filhos, lista_direcoes))
        filho_direcao_cont= list(reversed(filho_direcao))
        for filho, direcao in filho_direcao_cont:
            novo_atual = atual + [filho]
            novo_direcoes = direcoes + [direcao]
            pilha.append((novo_atual,novo_direcoes))
    return None

estados,movimentos=dfs(tabuleiro,bloco_final)

#print("Os estados foram: ")
#pprint(estados)

print("Os movimentos para chegar à solução foram: ", movimentos)
```

```python
#ALGORITMO DE PESQUISA GULOSA

import heapq

def greedy_search(estado_inicial, bloco_final, heuristica):
    start = time.time()
    fila = []
    heapq.heappush(fila, (heuristica(estado_inicial, tabuleiro), estado_inicial, [],[]))
    visitados = set()
    while fila:
        h, estado, caminho, direcoes = heapq.heappop(fila)
        if estado == bloco_final:
            end = time.time()
            print(f"Tempo: {round((end - start)*1000, 3)} ms")
            return caminho,direcoes
        if str(estado) in visitados:
            continue
        visitados.add(str(estado))
        lista_filhos, lista_direcoes = filhos(estado)
        for filho, direcao in zip(lista_filhos, lista_direcoes):
            heapq.heappush(fila, (heuristica(filho, tabuleiro), filho, caminho + [filho], direcoes + [direcao]))
    return None

estados,movimentos = greedy_search(tabuleiro, bloco_final, h1)

#print("Os estados foram: ")
#pprint(estados)

print("Os movimentos para chegar à solução foram: ", movimentos)
```
